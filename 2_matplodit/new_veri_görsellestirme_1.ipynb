```python
# 12 aylık bir borsa verisi temsil eden grafiği oluştur
#xticks ay isimleri olacak
# iki adet kripto para verisi olsun en düşük değer 1000 en yüksek  5000

dizi_1 = np.random.randint(1000,5000,12)
dizi_2 = np.random.randint(1000,5000,12)

uzunluk= np.arange(dizi_1.shape[0]) 

plt.plot(dizi_1,label="BTC")
plt.plot(dizi_2,label="LTC")

plt.xlabel("aylar")
plt.ylabel("Fiyat/TL")
plt.title("2024 Yılı Cripto Para")
plt.legend()

plt.xticks(uzunluk,["ocak","şubat","mart","nisan","mayıs","haziran","temmuz","agustos",
                    "eylül","ekim","kasım","aralık",], rotation=45)
plt.show()
```

```python
#bar çubuk grafikleri 1. gösterim: üst üste gösterim
fiyat_btc=np.random.randint(100,250,5)
fiyat_etc = np.random.randint(100,250,5)

uzunluk=np.arange(fiyat_btc.shape[0])

plt.bar(uzunluk, fiyat_btc,label="BTC")
plt.bar(uzunluk, fiyat_etc,label="ETC")
plt.xticks(uzunluk,["2020","2021","2022","2023","2024",])
plt.legend()
plt.show()
```

```python
#bar çubuk grafikleri 1. gösterim: yan yana gösterim
fiyat_btc=np.random.randint(100,250,5)
fiyat_etc = np.random.randint(100,250,5)

dizi_isim=["Bitcoin", "Ethereum"]
ort_btc= np.mean(fiyat_btc)
ort_etc= np.mean(fiyat_etc)

new_dizi=[ort_btc, ort_etc]

plt.bar(dizi_isim,new_dizi,color=["red","brown"],label=["BTC", "ETH"])

#plt.xticks(uzunluk,["2020","2021","2022","2023","2024",])
plt.legend()


for i in range(len(dizi_isim)):
    plt.text(i,new_dizi[i],f"{new_dizi[i]:.2f}",ha="center",va="bottom")

plt.show()
```

```python
# pie chart (posta grafiği)

giderler=[180,550,350,130,290]

plt.pie(giderler)
plt.show()
```

```python
giderler=[180,550,350,130,290]

etiketler=["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","IT"]
plt.pie(giderler,labels=etiketler)
plt.show()
```

```python
giderler=[180,550,350,130,290]
etiketler=["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","IT"]

ayirt_etme=[0,0,0,0.4,0]
#türkçe yüzde gösterimi---> autopct="%%%.1f"
plt.pie(giderler,labels=etiketler,autopct="%1.1f%%", shadow=True,explode=ayirt_etme)
plt.show()
```
