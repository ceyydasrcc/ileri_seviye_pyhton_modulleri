```python
#bar Ã§ubuk grafikleri 1. gÃ¶sterim: Ã¼st Ã¼ste gÃ¶sterim
fiyat_btc=np.random.randint(100,250,5)
fiyat_etc = np.random.randint(100,250,5)

uzunluk=np.arange(fiyat_btc.shape[0])

plt.bar(uzunluk, fiyat_btc,label="BTC")
plt.bar(uzunluk, fiyat_etc,label="ETC")
plt.xticks(uzunluk,["2020","2021","2022","2023","2024",])
plt.legend()
plt.show()
```

```python
#bar Ã§ubuk grafikleri 1. gÃ¶sterim: yan yana gÃ¶sterim
fiyat_btc=np.random.randint(100,250,5)
fiyat_etc = np.random.randint(100,250,5)

dizi_isim=["Bitcoin", "Ethereum"]
ort_btc= np.mean(fiyat_btc)
ort_etc= np.mean(fiyat_etc)

new_dizi=[ort_btc, ort_etc]

plt.bar(dizi_isim,new_dizi,color=["red","brown"],label=["BTC", "ETH"])

#plt.xticks(uzunluk,["2020","2021","2022","2023","2024",])
plt.legend()


for i in range(len(dizi_isim)):
    plt.text(i,new_dizi[i],f"{new_dizi[i]:.2f}",ha="center",va="bottom")

plt.show()
```
